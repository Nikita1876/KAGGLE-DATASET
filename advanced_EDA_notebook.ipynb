# 📊 Advanced EDA: Smartphone Usage, Behavior & Productivity

```
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('advanced_smartphone_behavior_dataset.csv')
df.head()
```

```
df.info()
```

```
df.describe()
```

## Distribution of Key Features

```
df['Daily_Screen_Time_Hours'].hist()
plt.title('Distribution of Screen Time')
plt.xlabel('Hours')
plt.ylabel('Frequency')
plt.show()
```

```
df['Sleep_Hours'].hist()
plt.title('Distribution of Sleep Hours')
plt.xlabel('Hours')
plt.ylabel('Frequency')
plt.show()
```

## Screen Time vs Addiction Level

```
df.boxplot(column='Daily_Screen_Time_Hours', by='Addiction_Level')
plt.title('Screen Time vs Addiction')
plt.suptitle('')
plt.xlabel('Addiction Level')
plt.ylabel('Screen Time')
plt.show()
```

## Sleep vs Productivity

```
plt.scatter(df['Sleep_Hours'], df['Productivity_Score'])
plt.xlabel('Sleep Hours')
plt.ylabel('Productivity Score')
plt.title('Sleep vs Productivity')
plt.show()
```

## Preferred App Usage

```
df['Preferred_App'].value_counts().plot(kind='bar')
plt.title('App Usage Distribution')
plt.xlabel('App')
plt.ylabel('Count')
plt.show()
```

## Correlation Matrix

```
corr = df.corr(numeric_only=True)
plt.imshow(corr)
plt.xticks(range(len(corr.columns)), corr.columns, rotation=45)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title('Correlation Matrix')
plt.colorbar()
plt.show()
```
