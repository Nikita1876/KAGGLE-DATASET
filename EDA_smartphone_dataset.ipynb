# 📊 EDA: Smartphone Usage & Addiction Dataset

```
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('advanced_smartphone_behavior_dataset.csv')
df.head()
```

```
df.info()
```

```
df.describe()
```

```
df['Addiction_Level'].value_counts()
```

```
plt.figure()
df.boxplot(column='Daily_Screen_Time_Hours', by='Addiction_Level')
plt.title('Screen Time vs Addiction Level')
plt.suptitle('')
plt.xlabel('Addiction Level')
plt.ylabel('Screen Time (Hours)')
plt.show()
```

```
plt.figure()
plt.scatter(df['Sleep_Hours'], df['Productivity_Score'])
plt.xlabel('Sleep Hours')
plt.ylabel('Productivity Score')
plt.title('Sleep vs Productivity')
plt.show()
```
